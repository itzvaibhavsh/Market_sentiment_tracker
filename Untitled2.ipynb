```python
!pip install textblob
from textblob import TextBlob
import pandas as pd

# Sample data (you can expand this later)
data = [
    {"date": "27/07/2025", "headline": "Gold climbs as dollar weakens", "price_next_day": 0.54},
    {"date": "27/07/2025", "headline": "Inflation fears push bond yields higher", "price_next_day": -0.30},
    {"date": "28/07/2025", "headline": "Oil prices recover after US inventory drop", "price_next_day": 0.65},
    {"date": "28/07/2025", "headline": "Rupee falls to all-time low", "price_next_day": -0.72},
    {"date": "29/07/2025", "headline": "Markets steady ahead of Fed decision", "price_next_day": 0.03},
]

results = []
for row in data:
    sentiment_score = TextBlob(row["headline"]).sentiment.polarity
    if sentiment_score > 0.1:
        sentiment = "Positive"
    elif sentiment_score < -0.1:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    results.append({
        "date": row["date"],
        "headline": row["headline"],
        "sentiment": sentiment,
        "price_next_day": row["price_next_day"]
    })

# Create and export DataFrame
df = pd.DataFrame(results)

# Save to Excel (will download as file)
df.to_excel("market_sentiment_tracker.xlsx", index=False)
df
```
